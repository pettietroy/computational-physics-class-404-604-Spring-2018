```python
#Method that solves the quadratic equation ax^2+bx+c = 0 with the traditional solution
#Input: Real numbers A, B, C, and list to store
#Output: NONE
from sympy import sqrt
from decimal import *
getcontext().prec = 40
def Quad(a, b, c, list_first):
    
    #Calculate the discriminant and its root
    d = (b**2) - (4*a*c)
    sqrt_d = sqrt(d).evalf(16)

    #Find two solutions
    sol1 = (-b-sqrt_d)/(2*a)
    sol2 = (-b+sqrt_d)/(2*a)
    
    list_first.append(sol1)
    list_first.append(sol2)

    return

#Method that solves the quadratic equation ax^2+bx+c = 0 using the second solution
#Input: Real numbers A, B, C, and list to store
#Output: NONE
def Quad_N(a, b, c, list_second):
    
    #Calculate the discriminant and its root
    d = (b**2) - (4*a*c)
    sqrt_d = sqrt(d).evalf(16)
    
    #Find two solution
    sol1 = (-2*c)/(b - sqrt_d)
    sol2 = (-2*c)/(b + sqrt_d)
    
    list_second.append(sol1)
    list_second.append(sol2)
    
    return 

#Main portion of the program
power_max = 7
base = 10.0
a = 1.0
b = 1.0

list_1st = [0]
list_2nd = [0]

#List of actual calculated values
list_actual = [0] * power_max*2
list_actual[1] = .9898979485566356
list_actual[2] = .010102051443364402
list_actual[3] = .998998997994986
list_actual[4] = .0010010020050140178
list_actual[5] = .9998999899979994
list_actual[6] = .00010001000200049459
list_actual[7] = .9999899998999979
list_actual[8] = .000010000100002016676
list_actual[9] = .999998999999
list_actual[10] = .000001000001000006634

dif_1st = [0] * power_max
dif_2nd = [0] * power_max

#Variables to keep track of better algorithm
winner1 = 0 
winner2 = 0

#Calculating Roots
for x in range(2,power_max):
    c = base**(-x)
    Quad(a, b, c, list_1st)
    Quad_N(a, b, c, list_2nd)
    
#Finding Differences
for x in range(0, power_max):
    dif_1st[x] = list_1st[x] - list_actual[x]
    dif_2nd[x] = list_2nd[x] - list_actual[x]
    
    #Magnitude of difference wins
    if(abs(dif_1st[x]) > abs(dif_2nd[x])):
        winner1 += 1
        
    if(abs(dif_1st[x]) < abs(dif_2nd[x])):
        winner2 += 2

if(winner1 > winner2):
        print("Algorithm One Is Better!")
        
if(winner1 < winner2):
        print("Algorithm Two Is Better!")
    

        
        
```
